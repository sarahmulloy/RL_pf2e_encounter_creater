```python
#example environment step 
def step(action):

    reward = 0

    if action == "up":
        player_x -= 1

    elif action == "attack":

        distance = abs(player_x - enemy_x) + abs(player_y - enemy_y)

        if distance == 1:
            reward += 20
            enemy_hp -= 20

    done = enemy_hp <= 0

    return get_state(), reward, done
```
